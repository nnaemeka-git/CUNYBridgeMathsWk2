# 1. Write a Python program to compute the eigenvalues and right eigenvectors of a given square array

```python
def eigenvector_value(matrx):
    if np.shape(matrx)[0] != np.shape(matrx)[1]:
        return ("Function takes only Square Matrices")
    else:
        eig_value,eig_vector = np.linalg.eig(np.array(matrx))
        print('Eigenvectors \n%s' %eig_vector)
        print('Eigenvalues \n%s' %eig_value)
    
    
```

Illustration

```python
A = np.random.randint(10, size=(3,3))
B = np.random.randint(10, size=(3,2))
```

```python
eigenvector_value(A)
```

```python
eigenvector_value(B)
```

# 3.  Write a Python program to compute the determinant of an array.

```python
def determinant(matrx):
    matrx = np.array(matrx)
    if np.shape(matrx)[0] != np.shape(matrx)[1]:
        return ("Function takes only Square Matrices")
    return round(np.linalg.det(matrx),4)
```
